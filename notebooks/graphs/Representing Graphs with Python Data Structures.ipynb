# Adjacency List Implementation
An adjacency list is a more efficient way of representing a sparsley populated graph compared to an adjacency matrix. At core an adjacency list is just an array of linked lists where:
- The index of the array represents a Vertex.
- The linked list associated with each index represents the other Verticies that are connected to the index Vertex.

At the core of most 'bootstrapped' graph implementations there are two objects that need to be written:
- A core object representing a single node: `Node`
- A class that contains the logic that is used to create, parse, add, remove and query nodes from the Graph Data Structure: `Graph`

For example the implementation of a Binary Search Tree was done with two main objects: 
```python
class TreeNode(object):
    pass

class BinaryTree(object):
    pass
```
Where the `TreeNode` object contains the logic for sotring Node associated values and the `BinaryTree` object contains the logic for correctly adding Nodes, performing searches and maintaining the Graph.

An adjacency list to represent a graph can be implemented in python very simply using python dictionaries:

````python
# Example of a adjacency list using python dict:
adj_lst = {
    0: {1: 5, 5: 2},
    1: {2: 4},
    2: {3: 9},
    3: {4: 7, 5: 3},
    4: {0: 1},
    5: {4: 8}
}

# Same adjacency list but formatted to be input into a networkx graph:
adj_lst_nx = {
    0: {1: {'weight': 5}, 5: {'weight': 2}},
    1: {2: {'weigth': 4}},
    2: {3: {'weight': 9}},
    3: {4: {'weight': 7}, 5: {'weight': 3}},
    4: {0: {'weight': 1}},
    5: {4: {'weight': 8}}
}
````

## This Adjacency List Represents the Graph that can be built via the Networkx Library:

````python
# Graph Object built from nested dicts:
G = nx.from_dict_of_dicts(adj_lst_nx, create_using=nx.Graph) 
G.edges.data('weight') # Telling networkx to look for weights under 'weight' key

# Plotting the Graph w/ Matplotlib:
nx.draw(G, with_labels=True, font_weight='bold')
plt.title('Example Graph Generated from Adjacency List w/ Networkx')
plt.show()
````

## Adjacency Matrix Implementation

I am much more familiar with representing graphs with matrix notation so a short example will be provided to familiarize myself with the networkx syntax and features:

````python
# Converting the existing adjacency list to an adjacency matrix:
adj_matrix = nx.adjacency_matrix(G, [0 ,1, 2, 3, 4, 5])
print(adj_matrix.todense())
print(type(adj_matrix.todense()))
````
